```python
pio = cv.imread("imgs/pio.jpeg")
assert pio is not None, "Image not found"
pio = cv.cvtColor(pio, cv.COLOR_BGR2RGB)

plt.imshow(pio)
```

```python
pioR = pio.copy()
pioR[:,:,1] = 0
pioR[:,:,2] = 0
plt.imshow(pioR)
```

```python
pioG = pio.copy()
pioG[:,:,0] = 0
pioG[:,:,2] = 0
plt.imshow(pioG)
```

```python
pioB = pio.copy()
pioB[:,:,0] = 0
pioB[:,:,1] = 0
plt.imshow(pioB)
```

```python
pio_glow = pio.copy()

glow = 2

for i in pio_glow:
    for j in i:
        #j[0] = max(min(j[0] * glow, 255), 0)
        #j[1] = max(min(j[1] * glow, 255), 0)
        j[2] = max(min(j[2] * glow, 255), 0)

plt.imshow(pio_glow)
```

```python
pio_glow = pio.copy()

for i in pio_glow:
    for j in i:
        j[0] = max(min(j[0] - 50, 255), 0)
        j[1] = max(min(j[1] - 50, 255), 0)
        j[2] = max(min(j[2] + 100, 255), 0)

plt.imshow(pio_glow)
```
